```python
import sqlite3
import pandas as pd
```

# Criar tabela.

```python
conn = sqlite3.connect('vendas.db')
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE vendas (
    id INTEGER PRIMARY KEY,
    produto TEXT,
    categoria TEXT,
    quantidade INTEGER,
    preco REAL,
    data TEXT
)
""")

conn.commit()
```

# Inserir dados

```python
dados = [
    ('Notebook', 'Eletrônicos', 2, 3500, '2024-01-10'),
    ('Mouse', 'Eletrônicos', 5, 80, '2024-01-12'),
    ('Teclado', 'Eletrônicos', 3, 150, '2024-01-15'),
    ('Cadeira', 'Móveis', 1, 900, '2024-01-20'),
    ('Mesa', 'Móveis', 1, 1200, '2024-01-22')
]

cursor.executemany("""
INSERT INTO vendas (produto, categoria, quantidade, preco, data)
VALUES (?, ?, ?, ?, ?)
""", dados)

conn.commit()
```

#Ver todas as vendas.

```python
pd.read_sql("SELECT * FROM vendas", conn)
```

#Total vendido por categoria.

```python
query = """
SELECT categoria,
       SUM(quantidade * preco) AS total_vendido
FROM vendas
GROUP BY categoria
"""
pd.read_sql(query, conn)
```

#Produto mais vendido (em valor).

```python
query = """
SELECT produto,
       SUM(quantidade * preco) AS faturamento
FROM vendas
GROUP BY produto
ORDER BY faturamento DESC
"""
pd.read_sql(query, conn)
```

#Vendas acima de R$ 1.000,00.

```python
pd.read_sql("""
SELECT * FROM vendas
WHERE quantidade * preco > 1000
""", conn)
```
